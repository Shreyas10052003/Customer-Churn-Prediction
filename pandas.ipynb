```python
#series is a 1 dimensional array can be created with tuple,list,dictionary
t=(10,11,12)
d=pd.Series(t) #'S' must be in capital 
print(d)
print(type(t))
```

```python
#creating a series using list
l=[25,286,365,43,87]
d=pd.Series(l)
d
```

```python
data=[98,27,55,41,39]
index=['a','b','c','d','e']
s=pd.Series(data,index)
print(s)
print(type(s))
```

```python
#creating series with array
#we don't need to import each time once we have imported at the starting then we can use those things anywhere
#now we are using "np" numpy
arr=np.array([1,2,3,4])
d3=pd.Series(arr)
d3
```

```python
#using dictionary
dict={'a':1,'b':2,'c':4,'d':5}
b=pd.Series(dict)
b
```
