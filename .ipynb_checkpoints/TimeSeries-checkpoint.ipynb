```python
#autocorrelation function
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data)
```

```python
from statsmodels.tsa.stattools import adfuller 
dftest = adfuller(data.Passengers,autolag='AIC')
print('1. ADF : ', dftest[0])
print('2. P-Value : ', dftest[1])
print('3. NUM of Lags : ', dftest[2])
print('4. num of observations used for ADF Regression and Critical Values Calculations : ', dftest[3])
print('5. Critical Values : ')
for key, val in dftest[4].items():
    print("\t",key,":",val)
#For this P-value is greater than 0.5 so accepting alternative hypothesis if less than 0.5 then accepting null hypothesis
#where it is non stationary data in this case.
```
