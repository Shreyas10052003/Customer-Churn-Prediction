```python
def classdiv(class1):
    if class1==1:
        return "business1"
    elif class1==2:
        return "business2"
    else:
        return "economic"
```

```python
data['class']=data['Pclass'].apply(classdiv)
data
```

```python
def classdiv1(Pclass,fare):
    if Pclass==1 and fare>30:
        return "luxury"
    elif Pclass==2 and fare>20:
        return "premium"
    elif Pclass==3:
        return "economy"
    else:
        return "basic"
```

```python
data['category1']=data[['Pclass','Fare']].apply(classdiv1)
data
```

```python
data['category']=data.apply(lambda row: classdiv1(row['Pclass'],row['Fare']),axis=1)
data
```
