```python
X=df2.iloc[:,:-1]
y=df2.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.60,random_state=42)
```

```python
y_train.value_counts()
```

```python
#SMOTE - SYNTHETIC MINORITY OVERSAMPLING TECHNIQUE -----BALANCING TECHNIQUE 
#it creates a synthetic duplicate data during training phase alone and avoid biasing during learning phase 
```

```python
from imblearn.over_sampling import SMOTE #imblearn --  imbalance learning 
smote=SMOTE(random_state=42)
X_train_smote,y_train_smote=smote.fit_resample(X_train,y_train)
```

```python
y_train_smote.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train_smote,y_train_smote)
```

```python
#STEP4 PREDICTON
y_pred=clf.predict(X_test)  #for testing phase you don't need to apply smote that is only for training phase
```

```python
from sklearn.metrics import *
accuracy_score(y_test, y_pred)
```
