```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')
```

```python
np.random.seed(42)
data=np.random.exponential(scale=2,size=1000)
df=pd.DataFrame(data,columns=['Values'])
```

```python
print("Skewness before transformation:",df['Values'].skew())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Values'],bins=30,kde=True)
plt.title("Before Transformation (Right-skewed Data)")
plt.show()
```

```python
df['Log_Transformed']=np.log1p(df['Values'])
print("Skewness after log transformation:",df['Log_Transformed'].skew())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Log_Transformed'],bins=30,kde=True)
plt.title("After Log Transformation")
plt.show()
```
